```python
results = pinstall([str(t) for t in df['tweet'].values])
results[:5], df.head()
```

```python
result_df = pd.DataFrame(results)
result_df
```

```python
# TODO: Merge dfs
# Save results in case kernel dies
result_df.to_csv('classifications.csv')
```

```python
all_df = df.merge(result_df, left_index=True, right_index=True)
all_df
```

```python
all_df[['tweet', 'label', 'score']]
```

```python
all_df['created_at'] = pd.to_datetime(all_df['created_at'])
all_df['year'] = all_df['created_at'].dt.year
```

```python
all_df.groupby('year')['label'].value_counts().plot.bar(color=['r', 'b', 'g'])
```

```python
import numpy as np
for label in ['pos', 'neg', 'neu']:
    all_df[f'weighted_{label}'] = np.where(all_df['label'] == label.upper(), all_df['score'], 0)
```

```python
all_df.groupby('year')[['weighted_pos', 'weighted_neg', 'weighted_neu']].sum().plot.bar(color=['g', 'r', 'b'])
```
